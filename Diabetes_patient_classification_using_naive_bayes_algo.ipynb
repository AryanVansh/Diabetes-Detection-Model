Importig necessary libraries

```python
import pandas as pd
import numpy as np
```

Creating Dataset from Dataframe

```python
df = pd.read_csv(r"/content/Complete-classification-algos-of-Supervised-Learning--main/Dataset-main/Diabetes.csv")
df.head
```

Analysing the data

```python
df.describe
```

All the columns in the dataset

```python
df.columns
```

Transfering all the data in 'x' dataframe

```python
x=df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age', 'diabetes']]
```

Droping the diabetes column

```python
x=df.drop("diabetes",axis=1)
x.head()
```

Saving the diabetese column in y dataset

```python
y=df[["diabetes"]]
y.head()
```

Splitting the dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,random_state=3000)
x_train
```

Changing the training data int vector form

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.fit_transform(x_test)
x_train
x_test
```

Training the model

```python
from sklearn.naive_bayes import GaussianNB
_=GaussianNB()
_.fit(x_train,y_train)
```

```python
y_pred=_.predict(x_test)
y_pred
```

Confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

Classification of report

```python
from sklearn.metrics import classification_report
classification_report(y_test,y_pred)
```
